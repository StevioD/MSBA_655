### 2. (20 pts) Modifying Data & Plotting with Dates
Using the weather data set (2021weatherbyMTcity.csv), find the maximum and minimum average temperatures for each day in January (city doesn't matter) and plot the max/min avg. temp lines by day for the month. Make sure the plot has labels for lines, axes, and create a title for it.

```python
import pandas as pd
pd.set_option('display.max_rows', None, 'display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
weather = pd.read_csv('2021weatherbyMTcity.csv')
weather.head(100)
```

```python
result_df = weather.groupby("Date").agg(max_temp = ("Avg_Temp", "max"), 
                                  min_temp = ("Avg_Temp", "min")).reset_index()

from matplotlib import pyplot as plt

plt.plot(result_df["Date"], result_df["max_temp"], label = "Maximum Temperature" )

plt.plot(result_df["Date"], result_df["min_temp"], label = "Minimum Temperature")

plt.title("Min-Maximum Temperature Records")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.xticks(rotation = 45)
plt.show()
```

### 3. (40 Points) Choosing the Best Visualization
There are 4 questions below. Using the Netflix dataset (neflix.csv), please answer the questions with a visualization and explain why you chose the chart type you did. (Make sure to include axis labels and plot titles)

```python
### Data 
netflix = pd.read_csv('netflix.csv')
netflix.head(100)
```

```python
#1. What are the top 5 Netflix Orginals based on IMDB Score?

netflix_top_5 = netflix.nlargest(5, ['IMDB Score'])

sns.set(rc={'figure.figsize':(15,10)})
sns.set_style('whitegrid')

ax = sns.barplot(x='IMDB Score', y='Title', data=netflix_top_5, palette='rocket')

ax.set_title('Top 5 IMDB Rated Netflix Movies', fontsize=20)
ax.set_xlabel('IMDB Score', fontsize=16)
ax.set_ylabel('', fontsize=16)
ax.set(xlim=(8, 9.2))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
#2. Which 3 Genre's have the highest average runtimes? 
netflix_genre_highrun = netflix.groupby(['Genre'], as_index=False).mean(['Runtime']).round(2).head(3)
netflix_genre_highrun

sns.set(rc={'figure.figsize':(12,8)})
sns.set_style('ticks')

ax = sns.barplot(x='Genre', y='Runtime', data=netflix_genre_highrun, color='lightblue')

ax.set_title('Top 3 Average Runtimes of Netflix Movies by Genre', fontsize=20)
ax.set_xlabel('Genre', fontsize=16)
ax.set_ylabel('Runtime in Minutes', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
#3. Do runtime & IMDB Score correlate to each other? (Please clarify your plot with an explanation of what you see.)
ax = sns.lmplot(x='IMDB Score', y='Runtime', data=netflix, height=10, line_kws={'color': 'red'})

# set the title, x label, and axis limits
plt.title('Scatter Plot of Runtime and IMDB Score', fontsize=20)
plt.xlabel('IMDB Score', fontsize=16)
plt.ylabel('Runtime in Minutes', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
# show heat map of the correlation between Runtime and IMDB Score
upp_mat = np.triu(netflix.corr())
sns.heatmap(netflix.corr(), vmin=-1,vmax=+1, annot=True, cmap='coolwarm', mask = upp_mat)
plt.show()
```

```python
# 4. Does IMDB score look different over time and is there any variation for English compared to the other languages? 
#    Hint: Consider making a new binary variable. 

def English(x):
    if x['Language'] == 'English':
        return "True"
    else:
        return "False"

netflix['Premiere'] = pd.to_datetime(netflix.Premiere)


netflix['English Language'] = netflix.apply(lambda x: English(x), axis=1)


netflix_eng_true = netflix.loc[netflix['English Language']=='True']
netflix_eng_true = netflix_eng_true.groupby(pd.Grouper(key='Premiere', freq="Y")).mean()
netflix_eng_true = netflix_eng_true.rename(columns={"IMDB Score":"IMDB Score True"})
netflix_eng_true = netflix_eng_true.drop('Runtime', axis=1)

netflix_eng_false = netflix.loc[netflix['English Language']=='False']
netflix_eng_false = netflix_eng_false.groupby(pd.Grouper(key='Premiere', freq="Y")).mean()
netflix_eng_false = netflix_eng_false.rename(columns={"IMDB Score":"IMDB Score False"})
netflix_eng_false = netflix_eng_false.drop('Runtime', axis=1)

netflix_eng = pd.merge(netflix_eng_true, netflix_eng_false, on='Premiere', how="outer")

netflix_eng
```

```python
sns.set_style("darkgrid")
mpg = sns.load_dataset('mpg')
mpg.head(100)
```

```python
#Step 2: Statistical analysis before proceeding with visualization.

#Get data type and non-null values of dataset.
mpg.info()

#Describe gives the statistical info about dataset.
mpg.describe()

#Check null values
mpg.isnull().any()

#Check uniqueness of horse power feature
mpg.horsepower.unique()
```

```python
#Here you can find "?" symbol. we can't allow horse power to have these type of characters, it needs to be removed.

#Replace "?" symbol to numpy nan.
mpg['horsepower'].replace("?",np.nan, inplace=True)

#Fill NA by using fillna() and replace it with median value.
mpg['horsepower'].fillna((mpg['horsepower'].median()), inplace=True)

#Now change the type to float.
mpg['horsepower'] = mpg['horsepower'].astype('float')
```

```python
#Step 3: Visualization:

#From above statistical exploration we can infer that, mpg is continuous variable , cylinders: is discrete variable , displacement: is continuous variable, horsepower: is continuous variable, weight: is continuous variable, acceleration: is continuous variable, model year: is discrete variable, origin: is discrete variable, car name: is a categorical variable.

#Explore the relationship between all the numerical values in the dataset in one plot.
#Pairplot library from seaborn gives the pairwise relationship between each variables.

#Here, we choose diagonal as "kde"
sns.pairplot(mpg,diag_kind='kde')
```

```python
#Observation from pair plot:
mpg.corr(method='kendall')
#From diagonal plots we can see origin has 3 points. model year shows two peaks majorily, acceleration & mpg are nearly normal, cylinder and displacement shows 3 clusters while weight shows 2 clusters. From bivariate plots we can see that mpg shows negative liner relationship with weight,horse power and displacement. Cylinder also shows negative correlation with levels.
```

```python
#Step 5:  Out of weight, horsepower, & cylinders, which variable appears to have the most influence on high MPG; use subplots to plot the relationship between the variables (at least 2 plots & there's not necessarily a correct answer here, more interested in logic behind it).

#Check the correlation for all parametes against each other before Visualization.

#corr() print the correlation in form of table.
corr = mpg.corr(method='kendall')
#Use Heatmap to visualize the correlation against each features.
plt.figure(figsize=(12,9))
sns.heatmap(corr, annot=True)
```

```python
#From above plot we infer correlation for,

#MPG and Weight =>-0.83, MPG and Horsepower=>-0.77 & MPG and Cylinder=>-0.78. This implies horsepower has high influence in MPG.

#Let us check the same with scatter plot also.

#A. Horsepower vs Mpg

sns.scatterplot(mpg['horsepower'],mpg['mpg'])
#Horsepower and Mpg are negatively correlated .

#B. Cylinder vs Mpg

sns.scatterplot(mpg['cylinders'],mpg['mpg'])

#Cylinder and Mpg have no correlation.

#C. Weight vs Mpg

sns.scatterplot(mpg['weight'],mpg['mpg'])

#Weight and Mpg are negatively correlated .
```

```python
#Step 6:  Chart depicting a relationship between a numerical & categorial variable.

#For chart analysis, we can either use bar chart or pie chart to depicting a relationship.

#Get the car names which has maximum mpg .
#Get maximum mpg
mpg['mpg'].max()

#maximum mpg will be 48.0. SO we can keep margin as 40.
car_names_40=mpg[mpg['mpg']>40]

#Plot bar char with origin as hue. It will differentiate the data with respect to origin.
sns.barplot(car_names_40['mpg'],car_names_40['name'],hue=car_names_40['origin'])
```
